# Storytelling Data Visualization on Exchange Rates

The dataset that we will be looking at note the exchange rates between the Euro, used by the European Union, and 41 other currencies. 

This dataset was put together by Daria Chemkaeva, and is available on Kaggle [here](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020)

```python
import pandas as pd

exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')

exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

There are a total of 5698 rows and 41 columns, with the rows corresponding with dates logged, and columns corresponding to a particular currency. 

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)

euro_to_dollar = exchange_rates[['Time', 'US_dollar']]

exchange_rates['US_dollar'].value_counts()
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'].astype(float)
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar['rolling_mean']
```

The idea that we are going to try to graph is how the euro-dollar rate changed under the last three US president's administrations. 

```python
import matplotlib.pyplot as plt
%matplotlib inline



bush_graph = euro_to_dollar[(euro_to_dollar['Time'].dt.year >= 2001) & 
                            (euro_to_dollar['Time'].dt.year < 2009)]
obama_graph = euro_to_dollar[(euro_to_dollar['Time'].dt.year >= 2009) & 
                            (euro_to_dollar['Time'].dt.year < 2017)]
trump_graph = euro_to_dollar[(euro_to_dollar['Time'].dt.year >= 2017) & 
                            (euro_to_dollar['Time'].dt.year < 2021)]


# plt.plot(list(bush_graph['Time']), list(bush_graph['rolling_mean']))
# plt.show()

plt.style.use('fivethirtyeight')

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3,
                                        figsize=(12,6))
ax1 = plt.subplot(2,3,1)
ax2 = plt.subplot(2,3,2)
ax3 = plt.subplot(2,3,3)
axes = [ax1, ax2, ax3]

ax1.plot(bush_graph['Time'], bush_graph['rolling_mean'], color='#BF5FFF')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '',
                     '2007', '', '2009'])
ax1.set_ylim(0.8, 1.7)
ax1.set_yticks([1.0, 1.2, 1.4, 1.6])
ax1.set_yticklabels(['1.0', '1.2','1.4', '1.6'])
ax1.text(731516.0, 1.92, 'BUSH', fontsize=18, weight='bold',
        color='#BF5FFF')
ax1.text(731196.0, 1.78, '(2001-2009)', fontsize=15, weight='bold', alpha=0.2)

ax2.plot(obama_graph['Time'], obama_graph['rolling_mean'], color='#ffa500')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '',
                     '2015', '', '2017'])
ax2.set_ylim(0.8, 1.7)
ax2.set_yticks([1.0, 1.2, 1.4, 1.6])
ax2.set_yticklabels(['1.0', '1.2','1.4', '1.6'])
ax2.text(734316.0, 1.92, 'OBAMA', fontsize=18, weight='bold',
        color='#ffa500')
ax2.text(734138.0, 1.78, '(2009-2017)', fontsize=15, weight='bold', alpha=0.2)

ax3.plot(trump_graph['Time'], trump_graph['rolling_mean'], color='#00B2EE')
ax3.set_xticklabels(['', '2017', '', '2018', '', '2019', '',
                     '2020', '', '2021'])
ax3.text(736796.0, 1.92, 'TRUMP', fontsize=18, weight='bold',
        color='#00B2EE')
ax3.text(736745.0, 1.78, '(2009-2017)', fontsize=15, weight='bold', alpha=0.2)

ax3.set_ylim(0.8, 1.7)
ax3.set_yticks([1.0, 1.2, 1.4, 1.6])
ax3.set_yticklabels(['1.0', '1.2','1.4', '1.6'])

ax1.text(730016.0, 2.35, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(730016.0, 2.14, '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)''',
        fontsize=16)

plt.show()


```
