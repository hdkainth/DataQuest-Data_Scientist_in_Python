# Exploring eBay Car Sales Data

This project looks to analyze a dataset of used cars from a "classifieds" section of the German eBay website called *eBay Kleinanzeigen*. The original dataset was scraped and uploaded to Kaggle. This version of the dataset was modified by DataQuest to sample 50,000 points as well as to simulate a scraped dataset.

The dataset has the following categories:

dateCrawled - When this ad was first crawled. All field-values are taken from this date.
name - Name of the car.
seller - Whether the seller is private or a dealer.
offerType - The type of listing
price - The price on the ad to sell the car.
abtest - Whether the listing is included in an A/B test.
vehicleType - The vehicle Type.
yearOfRegistration - The year in which which year the car was first registered.
gearbox - The transmission type.
powerPS - The power of the car in PS.
model - The car model name.
kilometer - How many kilometers the car has driven.
monthOfRegistration - The month in which which year the car was first registered.
fuelType - What type of fuel the car uses.
brand - The brand of the car.
notRepairedDamage - If the car has a damage which is not yet repaired.
dateCreated - The date on which the eBay listing was created.
nrOfPictures - The number of pictures in the ad.
postalCode - The postal code for the location of the vehicle.
lastSeenOnline - When the crawler saw this ad last online.

The dataset was originally scraped and uploaded to Kaggle by user orgesleka.

The original dataset isn't available on Kaggle anymore, but you can find it [here](https://data.world/data-society/used-cars-data).

```python
import numpy as np
import pandas as pd

autos = pd.read_csv('autos.csv', encoding='Latin-1')
print(autos.info())
print(autos.head())

```

## Primary Column Cleaning

We can see from looking at autos.info() and the information it presents is that most of the data is either an object type (string), or an int64 type. However, there are some values that should be registered as integers that are strings, namely price and odometer. Analyzing the first 5 rows, we see that this is because price has the '$' character, and odometer has the 'km' string attached to the numerical values.

```python
autos
```

We can also see that the names of the columns in the autos DataFrame are written in camel case, which isn't Python's preferred snake case. Along with that, some of the names could be better named. For example, notRepairedDamage can be renamed to unrepaired_damage. We will change the name of the columns and replace them with the snake case version. 

```python
print(autos.columns)
```

```python
new_columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'abtest', 'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model', 'odometer', 'registration_month', 'fuel_type', 'brand', 'unrepaired_damage', 'ad_created', 'num_pictures', 'postal_code', 'last_seen']
autos.columns = new_columns
autos.head()
```

We can use autos.describe() to find more info on the distribution of values in each column throughought the dataset. 

```python
autos.describe(include='all')
```

It's worth it to note that there are some columns that have close to one value throughout the whole data set, and these columns wouldn't add that much to the overall picture. The columns that have 2 unique values and a frequency of 49999 can be considered for being dropped. 

Here is where we see price and odometer's data being impacted through it being a string. We will go ahead and convert the column from a string type to an integer type.

```python
autos = autos.drop(['num_pictures', 'seller', 'offer_type'], axis=1)
```

```python
new_prices = []
prices = autos['price']
for price in prices:
    price = str(price)
    new_price = price.replace('$', '')
    new_price = new_price.replace(',', '')
    new_prices.append(new_price)
    
autos['price'] = new_prices
autos['price'] = autos['price'].astype(int)

new_odometers = []
current_odometers = autos['odometer']
for odometer in current_odometers:
    new_odometer = odometer.replace(',', '')
    new_odometer = new_odometer.replace('km', '')
    new_odometers.append(new_odometer)

autos['odometer'] = new_odometers
autos['odometer'] = autos['odometer'].astype(int)
autos.rename({'odometer': "odometer_km"}, axis=1, inplace=True)

    
```

```python
autos.describe(include='all')
```

However, looking at the quartiles for odometer, we can tell that they are perfectly rounded numbers to the decimal. Taking a look at the value counts can confirm this. 

```python
autos['odometer_km'].unique().shape
```

```python
autos['odometer_km'].value_counts()
```

```python
autos['odometer_km'].describe()
```

The distribution of the odometer column look normal, and there are no outliers that can be considered unrealistic, so we can leave the data be for now. However, while the odometer column can be fine, this doesn't guarantee that the data from other columns are correct. Now, we can check the price category that we modified earlier for outliers

```python
autos['price'].value_counts().sort_index(ascending=False).head(25)
```

We can see here that there are immediate outliers in the highest values, as a car price can't be 99 million dollars. If we look towards the bottom of our list, the price increases start to become realistic all the way until 350000, where afterwards, they start to jump to unrealistic numbers. Therefore, we will begin to parse our list from then on. 

```python
autos = autos[autos['price'].between(1, 350000)]
```

## Analyzing date columns

Analyzing the date values in date_crawled, ad_created, and last_seen, we see that it is stored in a string format, where the date and the time is listed back to back. The time is highly variable, so we can remove those values and focus on the dates to find common patterns. 

```python
autos['date_crawled'].str[:10].value_counts(normalize=True, dropna=False).sort_index()
        
```

We see here in date_crawled that the site was crawled faily from March 5th, 2016 to April 7th, 2016. Each date has close to the same number of entries on the table, around 0.03, or 3%. 

```python
autos['ad_created'].str[:10].value_counts(normalize=True, dropna=False).sort_index()
```

There aren't any patterns that we can glean from the ad_created column, other than the fact that some dates go back to June 2015.

```python
autos['last_seen'].str[:10].value_counts(normalize=True, dropna=False).sort_index()
```

A large amount of dates are the last three dates shown, April 5th to April 7th. This could potentially mean that a lot of cars were sold on these three days.

## Exploring brand data

We can analyze information based on the brands of cars that are being sold, as there is a "brand" column that we can analyze. In this case, we are going to analyze brands that have over 1% of entries.

```python
brand_percentages = autos['brand'].value_counts(normalize=True)
```

```python
brand_names = brand_percentages[brand_percentages > 0.01].index
```

```python
brand_mean_values = {}
for brand in brand_names:
    brand_values = autos[autos['brand'] == brand]
    brand_mean_values[brand] = brand_values['price'].mean()
```

In this instance, we will be attempting to find the mean prices of the brands listed. 

```python
brand_mean_values
```

```python
bmp_series = pd.Series(brand_mean_values)
brand_info = pd.DataFrame(bmp_series, columns=['mean_price'])
```

```python
brand_mean_mileage = {}
for brand in brand_names:
    brand_values = autos[autos['brand'] == brand]
    brand_mean_mileage[brand] = brand_values['odometer_km'].mean()
    
bmm_series = pd.Series(brand_mean_mileage)
brand_info['mean_mileage'] = bmm_series
brand_info
```

While the mileage generally stays pretty consistent, the most expensive cars tend to come from either bmw, mercedes benz, or audi.
